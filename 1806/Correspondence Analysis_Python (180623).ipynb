- 대응분석 : 자료의 행과 열범주를 저차원 공간상(2 차원)의 점들로 동시에 나타내어, 그들의 관계를 탐구하려는 탐색적 자료 분석 기법. <br>주성분분석과 비슷하지만, 대응분석은 범주형 데이터를 분석하는 방법이다.


- MCA(Multiple Correspondence Analysis) - 다중대응분석: 세 개 이상의 범주형 변수 간의 관계를 탐색하려면 다중 대응 분석을 사용합니다. 다중 대응 분석에서는 지시 변수의 행렬에 대해 단순 대응 분석을 수행합니다. 이 때 행렬의 각 열은 범주형 변수의 한 수준에 해당합니다. 단순 대응 분석의 이원 분할표를 평가하는 대신 다원 분할표를 1차원으로 축약합니다. 다중 대응 분석 절차에서는 단순 대응 분석에 비해 잠재적으로 더 많은 수의 변수에 대한 정보를 얻을 수 있지만 행과 열의 관계에 대한 정보는 잃을 수도 있습니다.




```python
import pandas as pd
from mca import MCA as MCA
import numpy as np

%cd C:\Users\hanbum\Desktop
```

```python
data = pd.read_csv('burgundies.csv', skiprows=1, index_col=0, header=0)

data
```

```python
X = data.drop('oak_type', axis=1)
X
```

```python
j_sup = data.oak_type
i_sup = np.array([0, 1, 0, 1, 0, .5, .5, 1, 0, 1, 0, 0, 1, 0, .5, .5, 1, 0, .5, .5, 0, 1])
ncols = 10
```

```python
mca_ben = MCA(X, ncols=ncols)
mca_ind = MCA(X, ncols=ncols, benzecri=False)
data = {'Iλ': pd.Series(mca_ind.L),
        'τI': mca_ind.expl_var(greenacre=False, N=4),
        'Zλ': pd.Series(mca_ben.L),
        'τZ': mca_ben.expl_var(greenacre=False, N=4),
        'cλ': pd.Series(mca_ben.L),
        'τc': mca_ind.expl_var(greenacre=True, N=4)}

# 'Indicator Matrix', 'Benzecri Correction', 'Greenacre Correction'
columns = ['Iλ', 'τI', 'Zλ', 'τZ', 'cλ', 'τc']
table2 = pd.DataFrame(data=data, columns=columns).fillna(0)
table2.index += 1
table2.loc['Σ'] = table2.sum()
table2.index.name = 'Factor'

table2
```
