```python
import pandas as pd
import numpy as np
```

```python
#base되는 data frame 생성

d = {'House':['Stark','Targaryen','Lannister','Baratheon','Tyrell'],'Region':['The North',"Slaver's Bay",'The Westerlands'
                                                                             ,'The Stormlands','The Reach']}

e = {'House':['Tully','Greyjoy'], 'Region':['The Riverlands','The Iron Islands']}

house = pd.DataFrame(d)
house_extra = pd.DataFrame(e)
```

```python
house
```

```python
house_extra
```

```python
# house에 house_extra 붙이기(밑으로, 열추가하기)

house = pd.concat([house,house_extra],axis = 0)
```

```python
house
```

```python
#행 추가 해보기

house.loc[len(house)] = ['Redwyne','The Reach']
```

```python
# merge --> left / right / inner / outer

d = {'House':['Lannister','Stark','Stark','Lannister','Targaryen','Baratheon','Mormont'],
     'Name':['Jamie Lannister','Robb Stark','Arya Stark','Cersi Lannister','Daenerys Targaryen','Robert Baratheon','Jorah Mormont']}

e = {'House':['Stark','Targaryen','Lannister','Baratheon','Tyrell'],
     'Region':['The North',"Slaver's Bay",'The Westerlands','The Stormlands','The Reach'],
    'Military_Strength':['20000','110000','60000','40000','30000']}

candidates = pd.DataFrame(d)
house = pd.DataFrame(e)
```

```python
house5 = pd.merge(candidates, house, on = "House", how='left')
```

```python
house5
```

```python
house2 = pd.merge(candidates, house, on = "House", how = 'right')
```

```python
house3 = pd.merge(candidates, house, on='House', how = 'inner')
```

```python
house4 = pd.merge(candidates, house, on = "House", how = 'outer')
```

```python
# suffixes

d = {'House':['Lannister','Stark','Stark','Lannister','Targaryen','Baratheon','Mormont'],
     'Name':['Jamie Lannister','Robb Stark','Arya Stark','Cersi Lannister','Daenerys Targaryen','Robert Baratheon','Jorah Mormont'],
    'Region':['Westros','North','Westros','Westros','Mereene','Westros','Mereene']}

e = {'House':['Stark','Targaryen','Lannister','Baratheon','Tyrell'],
     'Region':['The North',"Slaver's Bay",'The Westerlands','The Stormlands','The Reach'],
    'Military_Strength':['20000','110000','60000','40000','30000']}

candidates = pd.DataFrame(d)
house = pd.DataFrame(e,columns=['House','Region','Military_Strength'])
```

```python
house_suffixes = pd.merge(candidates, house, on ='House', how='left', suffixes=('_left','_right'))
```

```python
house_suffixes
```

```python
# removing duplicate ---> 새로운 dataframe에 저장 못하나???

candidates.drop_duplicates(subset=['House'], keep = 'first',inplace=True)
```

```python
house_dupli = pd.merge(candidates, house, on='House', how ='left')
```

```python
house_dupli
```

```python
# remove column

house_suffixes = house_suffixes.drop('Region_right', axis=1)
```

```python
# change value in data frame

house_suffixes[house_suffixes['Name']=='Arya Stark'] = house_suffixes[house_suffixes['Name']=='Arya Stark'].replace('Westros','North')
```

```python
house_suffixes
```

```python
#change all value in dataframe : inplace --> 원본 dataframe 자체를 변경

house_suffixes.replace('North','King',inplace =True)
```

```python
# change the name of the column

house_suffixes2 = house_suffixes.copy()


house_suffixes2.columns = house_suffixes2.columns.str.replace('Region_left','Region')
```
